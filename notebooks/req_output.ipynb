```python
url = 'https://restcountries.com/v3.1/alpha/840'
```

```python
import requests
```

```python

# make an HTTP GET request to a server
response = requests.get(url)

if response.status_code == 200:
    try:
        data = response.json()
    except:
        print('JSON not found')
        data = response.text
    print(data)
```
